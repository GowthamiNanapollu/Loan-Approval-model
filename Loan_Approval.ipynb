## Data Cleaning

#### check for null data

```python
data.isnull().sum()
```

```python
data.info()
```

```python
age_q = data.person_age.quantile(0.9)
income_q = data.person_income.quantile(0.9)
exp_q = data.person_emp_exp.quantile(0.9)
loan_q = data.loan_amnt.quantile(0.9)
rate_q = data.loan_int_rate.quantile(0.9)

credit_q = data.cb_person_cred_hist_length.quantile(0.9)
cscore_q = data.credit_score.quantile(0.9)
```

##### Outlier removing

```python
data = data[

    (data.person_age<= age_q) & 
    (data.person_emp_exp	 <= exp_q )&	
    (data.loan_amnt <= loan_q )&	
    (data.loan_int_rate <= rate_q) &	
    (data.cb_person_cred_hist_length	 <= credit_q) &
    (data.credit_score <= cscore_q)
]
```

```python
data= data.reset_index(drop= True)
```

```python
data.shape
```

```python
data.describe(include ='all')
```

```python
data.info()
```

#### Mapping for Gender, Previous Loan defaults Attribute

```python
data['person_gender'] = data.person_gender.map({'female' : 0, 'male' : 1})
data
```

```python
data['previous_loan_defaults_on_file'] = data.previous_loan_defaults_on_file.map({'Yes': 1, 'No': 0})
```

#### Creating dummies for Education, Home ownership, Loan Intent

```python
education = pd.get_dummies(data.person_education, drop_first = True)
education = education.astype(int)
education
```

```python
home = pd.get_dummies(data.person_home_ownership, drop_first = True)
home = home.astype(int)
home
```

```python
intent = pd.get_dummies(data.loan_intent, drop_first= True)
intent = intent.astype(int)
intent
```

```python
dummies = pd.concat([education,home, intent], axis= 1)
dummies
```

```python
cleaned_data = data.drop(['person_education','person_home_ownership','loan_intent'], axis= 1)
```

```python
input_without_dummies = cleaned_data.copy()
```

```python
input_without_dummies = input_without_dummies.drop('loan_status', axis =1)
```

```python
input_without_dummies
```

## Standardizing inputs 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(input_without_dummies)
```

```python
scaled_inputs_without_dummies = scaler.transform(input_without_dummies)
```

```python
dummies.to_numpy()
```

```python
inputs = np.concatenate([scaled_inputs_without_dummies, dummies], axis= 1)
```

```python
inputs.shape
```

```python
targets  = cleaned_data['loan_status']
```

```python
targets
```

## Spliting Data

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest,ytrain, ytest = train_test_split(inputs, targets, test_size= 0.2, random_state = 42)
```

```python
ytrain
```

## Training a Logistic Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain, ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
Loan_model = pd.DataFrame()
```

```python
Loan_model['Features'] = [
    'person_age',
    'person_gender',
    'person_income',
    'person_emp_exp',
    'loan_amnt',
    'loan_int_rate',
    'loan_percent_income',
    'cb_person_cred_hist_length',
    'credit_score',
    'previous_loan_defaults_on_file',
    'Bachelor',
    'Doctorate',
    'High School',
    'Master',
    'OTHER',
    'OWN',
    'RENT',
    'EDUCATION',
    'HOMEIMPROVEMENT',
    'MEDICAL',
    'PERSONAL',
    'VENTURE'

]
```

```python
 Loan_model['Weights'] =np.transpose(model.coef_)
```

```python
intercept = pd.DataFrame([{'Features' : 'intercept', 'Weights': -5.400372449}])
```

```python
Loan_model = pd.concat([intercept,Loan_model])
```

```python
Loan_model
```

```python
Loan_model['Odds'] = np.exp(Loan_model.Weights)
```

#### Model Coffecients

```python
Loan_model
```

## Predicting the Output

```python
yhat = model.predict(xtest)
```

```python
yhat
```

## Manually evaluate Yhat

```python
correct_pred = 0
for i in range(len(yhat)):
    if yhat[i] == ytest.iloc[i]:
        correct_pred += 1
```

```python
accuracy = correct_pred/ yhat.shape[0]
accuracy
```
